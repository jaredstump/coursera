```python
import sys
import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
import traceback
```

```python
def date_time(table_cells):
    return [data_time.strip() for data_time in list(table_cells.strings)][0:2]
def booster_version(table_cells):
    out = ''.join([booster_version for i,booster_version in enumerate(table_cells.strings) if i%2==0][0:-1])
    return out
def landing_status(table_cells):
    out = [i for i in table_cells.strings][0]
    return out
def get_mass(table_cells):
    mass = unicodedata.normalize("NFKD", table_cells.text).strip()
    if mass:
        mass.find("kg")
        new_mass = mass[0:mass.find("kg")+2]
    else:
        new_mass = 0
    return new_mass
def extract_column_from_header(row):
    if (row.br):
        row.br.extract()
    if row.a:
        row.a.extract()
    if row.sup:
        row.sup.extract()
    column_name = ' '.join(row.contents)
    if not(column_name.strip().isdigit()):
        column_name = column_name.strip()
        return column_name
```

```python
static_url = "https://en.wikipedia.org/w/index.php?title=List_of_Falcon_9_and_Falcon_Heavy_launches&oldid=1027686922"
data = requests.get(static_url).text
soup = BeautifulSoup(data)
print(soup.title)
html_tables = soup.find_all('table')[2].find_all('tr')[0].find_all('th')
```

```python
column_names = []
for row in html_tables:
    column_names.append(extract_column_from_header(row))
column_names
```

```python
launch_dict = dict.fromkeys(column_names)
del launch_dict['Date and time ( )']
launch_dict['Flight No.'] = []
launch_dict['Launch site'] = []
launch_dict['Payload'] = []
launch_dict['Payload mass'] = []
launch_dict['Orbit'] = []
launch_dict['Customer'] = []
launch_dict['Launch outcome'] = []
launch_dict['Version Booster'] = []
launch_dict['Booster landing'] = []
launch_dict['Date'] = []
launch_dict['Time'] = []
```

```python
launch_dict
```

```python
extracted_row = 0
try:
    for table_number, table in enumerate(soup.find_all('table', 'wikitable plainrowheaders collapsible')):
        for row_number, rows in enumerate(table.find_all('tr')):
            if rows.th:
                if rows.th.string:
                    flight_number = rows.th.string.strip()
                    flag = flight_number.isdigit()
            else:
                flag = False
            cells = rows.find_all('td')
            if flag:
                extracted_row += 1
                launch_dict['Flight No.'].append(extracted_row)
                datatimelist = date_time(cells[0])
                date = datatimelist[0].strip(',')
                launch_dict['Date'].append(date)
                time = datatimelist[1]
                launch_dict['Time'].append(time)
                bv = booster_version(cells[1])
                if not(bv):
                    bv = cells[1].a.string
                launch_dict['Version Booster'].append(bv)
                launch_site = cells[2].a.string
                launch_dict['Launch site'].append(launch_site)
                payload = cells[3].a.string
                launch_dict['Payload'].append(payload)
                payload_mass = get_mass(cells[4])
                launch_dict['Payload mass'].append(payload_mass)
                orbit = cells[5].a.string
                launch_dict['Orbit'].append(orbit)
                #customer = cells[6].a.string
                #customer = cells[6].string
                if cells[6] and cells[6].a:
                    customer = cells[6].a.string
                else:
                    customer = cells[6].string
                launch_dict['Customer'].append(customer)
                #outcome = cells[7].string.strip()
                outcome = list(cells[7].strings)[0]
                launch_dict['Launch outcome'].append(outcome)
                booster_status = landing_status(cells[8])
                launch_dict['Booster landing'].append(booster_status)
except AttributeError as e:
    print(f"{e}")
    traceback.print_exc()
    #print(f"error on {extracted_row},  \n {flag} \n {cells}")
    print(f"table number {table_number}")
    print(f"row number {row_number}")
    #print(f"table: \n {table}")
   #print(f"row: \n {rows}")
    print(f"cell \n {rows.find_all('td')[6]}")
    #print(f"cell: \n {cells[6]}")
```
